```python
from scipy.stats import ttest_rel
import numpy as np

#H0 no change in marks
#H1 change in marks

before_marks = np.array([23,3,45,23,22,6,43,22,44])
after_marks = np.array([66,76,55,88,78,98,89,88,76])

```

ttest_ind is for independant values 
ttest_rel is for dependant values i.e. same person ko diff marks like in the first example
ttest_1samp is for 1 sample testing

```python
t_test, p_val = ttest_rel(before_marks,after_marks)
print(f'The P-value is: {p_val}')

if p_val>0.05:
    print('There cannot be any change in the marks.')
else: 
    print('There can be change in marks.')
```

```python
from scipy.stats import ttest_1samp
marks = [23,3,45,23,22,6,43,22,44]
mean = np.mean(marks)

t_test, p_val = ttest_1samp(marks, 30)
print(f'P-value is: {p_val}')

if p_val>0.05:
    print('We can accept the null hypothesis.')
else: 
    print('We can reject the null hypothesis.')
```

```python
ai1 = [23,39,45,23,22,36,43,22,44]
ai2 = [3,34,4,2,21,6,42,22,4]
```

```python
from scipy.stats import ttest_ind
t_test, p_val = ttest_ind(ai1,ai2)
print(f'The P-value is: {p_val}')
if p_val>0.05:
    print('No diff.')
else: 
    print('There is diff.')
```
