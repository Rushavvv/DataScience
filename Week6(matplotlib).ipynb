```python
religion = ['hindu', 'buddhist', 'muslim']
population = [25,30,21]
```

```python

df['mean'] = (df['G1'] + df['G2'] + df['G3'])/3
df['Status'] = df['mean'].apply(lambda y:'Yes' if y>=8 else 'No')
```

```python
df
```

```python
import matplotlib as plt
df['Status'].value_counts()
```

```python
df[df['Status'] == 'Yes'].groupby('school')['Status'].value_counts().plot(kind = 'bar')
```

```python
df[df['Status'] == 'No'].groupby('studytime')['Status'].value_counts().plot(kind = 'bar')
```

```python
df['G1'].hist(bins = [6,8,11,13], ec = 'red')
```

```python
df['G1'].hist() 
df['G2'].hist()
plt.xlabel('Frequency of students')
plt.ylabel('Marks')
```

```python
numerical_df = df.select_dtypes(include = 'number')
for column in numerical_df.columns:
    print(f'\n Statistical data of {column}: ')
    print(f' The Sum is {numerical_df[column].sum()}')
    print(f' The Mean is {numerical_df[column].mean()}')
    print(f' The Standard Deviation is {numerical_df[column].std()}')
    print(f' The Skewness is {numerical_df[column].skew()}')
    print(f' The Kurtosis is {numerical_df[column].kurt()}')
```
