```python
# Importint ftest library
from scipy.stats import f_oneway
# H0: Response time across complaints is similar
# H1: Response time is not similar across complaints
# Getting unique complaint types 
complaintTypes = df['Complaint Type'].unique()
# Creating arrays for closing time for each complaint type 
group = []
for complaint in complaintTypes:
    time = df[df['Complaint Type'] == complaint]['Request_Closing_Time'].values
    group.append(time)
```

```python
# Running ftest 
f_test, p_value = f_oneway(*group)

print(f'The value of the ftest is {f_test} and the p value is {p_value}')

# Using p-value to see if Null Hypothesis is accepted of the Alternate Hypothesis is accepted. 
if p > 0.05: 
    print('Response time is similar across complaints \n H0 is accepted')
else: 
    print('Response time across complaints are different \n H0 is rejected and H1 is accepted')
```

```python
#Importing chi2 test library
from scipy.stats import chi2_contingency
# H0: There is no relationship between Complaint Type and Location Type
# H1: There is relationship between Complaint Type and Location Type 
testArray = pd.crosstab(df['Complaint Type'], df['Location Type'])
```

```python
# Performing the chi2 test
chi2, p, dof, expected = chi2_contingency(testArray)

# Displaying the values for chi2 and p value
print(f' The value of the chi2 is {chi2} \n The p-value is {p_value}')
# Using p-value to see if Null Hypothesis is accepted of the Alternate Hypothesis is accepted. 
if p > 0.05: 
    print('There is no relationship between Complaint Type and Location Type \n H0 is accepted')
else: 
    print('There is relationship between Complaint Type and Location Type \n H0 is rejected and H1 is accepted')
```
